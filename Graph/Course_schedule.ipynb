## 207. Course Schedule

### Descrption

There are a total of ```numCourses``` courses you have to take, labeled from 0 to ```numCourses - 1```. You are given an array prerequisites where prerequisites$[i] = [a_i, b_i]$ indicates that you must take course $b_i$ first if you want to take course $a_i$.

For example, the pair [0, 1], indicates that to take course 0 you have to first take course 1.
Return true if you can finish all courses. Otherwise, return false.

### Example

**Input**: numCourses = 2, prerequisites = [[1,0]]

**Output**: true

**Explanation**: There are a total of 2 courses to take. 
To take course 1 you should have finished course 0. So it is possible.


**Input**: numCourses = 2, prerequisites = [[1,0],[0,1]]

**Output**: false

**Explanation**: There are a total of 2 courses to take. 
To take course 1 you should have finished course 0, and to take course 0 you should also have finished course 1. So it is impossible.

### Solution

**Depth First Search**

We can model this problem to a directed graph, with each course as the vertex and relation between courses as edges, ie, if 0 is the prerequsite of course 1, then there is an edge that starts from 0 and points to 1. For example, suppose the input array looks like this:

```
[
 [1, 2]
 [1, 3]
 [2, 3]
 [2, 4]
 [3, 4]
]
```

The resulting graph is:

```
0 -> []
1 -> [2, 3]
2 -> [3, 4]
3 -> [4]
4 -> []
```

. In order to have a valid schedule, there must not be any confliction between course and its prerequisites. In other words, no cycle in the DAG. That's where DFS comes in, recall that DFS can be used to detect cycle in a directed graph. 

1. Build the graph using input array.

2. Initialize a dictionary to store status of each vertex.

3. For each vertex in the graph, label them as unvisited in the list built in previous step.

4. Run DFS on each vertex in the graph.

5. In each DFS call, if the vertex has been finished, then we return; if the vertex is unvisited, mark it as 'visited' and run DFS on it; if it's visited, then we have a cycle, return False. Then we mark the input vertex as 'finished'. 

````python
import collections

def canFinish(numCourses, prerequisites):
    
    #Construct the graph
    g = collections.defaultdict(list)
    for course, prereq in prerequisites:
        g[course].append(prereq)
        
    visited = [-1 for _ in range(numCourses)]
    for i in range(numCourses):
        if not DFS(g, i, visited):
            return False
    
    return True
    
def DFS(g, src, visited):
    if visited[src]==1:
        return True
    
    visited[src] = 0
    for v in g[src]:
        if g[src]==0:
            return False
        DFS(s, v, visited)
        
    visited[s] = 1
    return True
````
