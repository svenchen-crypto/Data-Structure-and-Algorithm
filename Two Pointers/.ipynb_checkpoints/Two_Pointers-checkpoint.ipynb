# Two Pointers

The two-pointer technique is a fundamental algorithmic approach used to solve real-world problems. Unlike strategy such as “sliding window,” which follow a specific pattern, or data structures like binary trees or graphs, two pointers is not a rigid “recipe.” Instead, it is a flexible strategy that involves using two (or more) pointers to traverse one or more iterable structures (e.g., arrays, linked lists, or strings) while maintaining a **“certain structure”** to solve the problem. Identifying this “structure” is crucial for the success of the two-pointer approach. However, since the two-pointer technique is not a predefined recipe, such a structure may not always exist. In this notebook, we will explore both scenarios. Before diving in, it’s worth noting that there are multiple variations of the two-pointer technique:

1. Traverse from start of the list

```
p1       p2
 ↓        ↓
 --- --- --- --- --- --- --- --- ---
| 2 | 1 | 2 | 0 | 1 | 0 | 1 | 0 | 1 |
 --- --- --- --- --- --- --- --- ---
```

2. Traverse from start and end of the list

```
left                            right
 ↓                                ↓
 --- --- --- --- --- --- --- --- ---
| 2 | 1 | 2 | 0 | 1 | 0 | 1 | 0 | 1 |
 --- --- --- --- --- --- --- --- ---
```

3. Traverse two iterables from start

```
p1                           
 ↓                                
 --- --- --- --- --- --- --- --- ---
| 2 | 1 | 2 | 0 | 1 | 0 | 1 | 0 | 1 |
 --- --- --- --- --- --- --- --- ---
 
 p2                            
 ↓
  --- --- --- --- --- --- --- --- ---
| 2 | 1 | 0 | 0 | 1 | 0 | 3 | 0 | 1 |
 --- --- --- --- --- --- --- --- ---
```

3. Fast&Slow Pointers

```
               slow               fast
                 ↓                 ↓
0 -> 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7
 
```
